```python
termsFile = open("../data/topnineterms.tsv")
inputrecord = {}
for line in termsFile:
    elementsOfLine = line.split("\t")
    year = elementsOfLine[0][:4]
    termList = elementsOfLine[2].split(" ")[:-1]
    if year not in inputrecord:
        inputrecord[year] = {}
    for term in termList:
        if term not in inputrecord[year]:
            inputrecord[year][term] = 1
        else:
            inputrecord[year][term]+=1
print inputrecord
```
